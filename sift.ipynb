```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('/home/marco/Pictures/Lena.png')
img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.figure(figsize=(5, 5))
plt.imshow(img, 'gray')
plt.show()
```

```python
sigma = 1.6
g1 = cv.getGaussianKernel(ksize=13, sigma=sigma)
g2 = cv.getGaussianKernel(ksize=13, sigma=(sigma + .5))
dog = g2 - g1
dog = dog * 1/np.sum(dog)

fig, axs = plt.subplots(1, 3)
axs[0].imshow(np.matmul(g1, g1.T), 'gray')
axs[1].imshow(np.matmul(g2, g2.T), 'gray')
axs[2].imshow(np.matmul(dog, dog.T), 'gray')
plt.show()
```

```python
img_dog = cv.sepFilter2D(img, cv.CV_32F, dog, dog.T)
plt.imshow(img_dog, 'gray')
plt.show()
```
